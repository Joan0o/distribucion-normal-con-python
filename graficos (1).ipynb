```python
from matplotlib import pyplot as plt
from scipy.stats import norm
import numpy as np
import pandas as pn
```

```python
#%matplotlib notebook
```

```python
_10 = pn.read_excel('Datos1.xlsx')["Diametro(mm)"]

std = np.std(_10)
mean = _10.sum()/len(_10)
```

```python
domain = np.linspace(_10.min(), _10.max())
plt.plot(domain, norm.pdf(domain, mean, std),  
         label=r'media='+str(mean)+' \ndesviación='+str(std))
plt.hist(_10, edgecolor='black', alpha=.5, density=True, range=(_10.min(), _10.max()), bins=10)
plt.title('Ajuste normal 10 arvejas')
plt.xlabel('Diametro')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
_200 = pn.read_excel('Datos2.xlsx')["Diametro(mm)"]

std = np.std(_200)
mean = _200.sum()/len(_200)
```

```python
domain = np.linspace(_200.min(), _200.max())
plt.plot(domain, norm.pdf(domain, mean, std),  
         label=r'media='+str(mean)+' \ndesviación='+str(std))
plt.hist(_200, edgecolor='black', alpha=.5, density=True, range=(_200.min(), _200.max()), bins=20)
plt.title('Ajuste normal 200 arvejas, 20 intervalos')
plt.xlabel('Diametro')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
_parametros = []
_grupo = []
for k, x in _200.items():
    _grupo.append(x)
    if ((k+1) % 10) == 0:
        mean = np.mean(_grupo)
        std = np.std(_grupo)
        _parametros.append({'mean': mean, 'std': std, 'data': _grupo})
        _grupo = []
```

```python
promedio_de_promedios = np.mean([x['mean'] for x in _parametros])
```
